7. What is the difference between "==" and "is" operators in python ?
  - "==" Operator: Equality
      The == operator checks if the values of two objects are equal. It compares the contents or data stored within the objects.

      Use Case:
        Compares the values or data of the objects.Returns True if the contents are the same.

      Example:
      list1 = [1, 2, 3]
      list2 = [1, 2, 3]
      print(list1 == list2)  # Output: True (values are the same)

      "is" operator checks if two objects are the same object in memory. It compares their memory locations or identities.

      Use Case:
        Compares the memory addresses of the objects.
        Returns True if both variables point to the same object in memory.
        Example:
        # Comparing two lists with the same content
        list1 = [1, 2, 3]
        list2 = [1, 2, 3]
      print(list1 == list2)  # Output: True (values are the same)
      print(list1 is list2)  # Output: False (different memory locations)

    Operator	Checks	              Example	Result
    ==	      Value equality	      [1, 2] == [1, 2]	True
    is	      Identity equality	    [1, 2] is [1, 2]	False
    is	      Identity equality	    x = 2; x is 2	True
    is	      Identity equality	    x = [1,2],y=[1,2] x is y-> false
    ==        Value equality        x = [1,2],y=[1,2] x is y-> True

8. What are logical operators in Python?
  - ogical operators in Python are used to combine multiple conditions or expressions and return a Boolean value (True or False) based on the logic of the expressions. These operators are typically used in conditional statements like if, while.

  Python has three logical operators:
    a. and
    b. or
    c. not

    Examples:
    and : The and operator returns True if both conditions are true; otherwise, it returns False
      x = 5
      y = 10
      print(x > 0 and y > 0)  # True (both conditions are True)
      print(x > 0 and y < 0)  # False (second condition is False)

      x = 5
      y = -10
      print(x > 0 or y > 0)  # True (first condition is True)
      print(x < 0 or y > 0)  # False (both conditions are False)

    not:The not operator is a unary operator that inverts the truth value of the condition.
      x = 5
      print(not (x > 0))  # False (x > 0 is True, so not makes it False)
      print(not (x < 0))  # True (x < 0 is False, so not makes it True)

    Combining Logical Operators
      x = 5
      y = 10
      z = -3

      # Combining and/or operators
      result = (x > 0 and y > 0) or z > 0
      print(result)  # True (x > 0 and y > 0 is True, so the entire condition is True)

      # Using not
      result = not (x > 0 and z < 0)
      print(result)  # False (x > 0 and z < 0 is True, but not inverts it)

10. What is the difference between implicit and explicit type casting ?
    -a.Implicit Type Casting
          Python automatically converts one data type to another when it is safe to do so.Typically occurs in operations where Python promotes smaller data types (like integers) to larger ones (like floats) to prevent data loss.
    Example:
      # Implicit type casting
      x = 10    # Integer
      y = 2.5   # Float
      result = x + y  # Integer is converted to float
      print(result)  # Output: 12.5
      print(type(result))  # Output: <class 'float'>
    
    b. Explicit Type Casting
        The programmer explicitly converts one data type to another using type conversion functions. Commonly used when data needs to be in a specific format.

        Examples of Explicit Type Casting:
        Converting Strings to Integers/Floats:
        str_num = "123"
        int_num = int(str_num)  # Converts string to integer
        float_num = float(str_num)  # Converts string to float
        print(int_num, type(int_num))  # Output: 123 <class 'int'>
        print(float_num, type(float_num))  # Output: 123.0 <class 'float'>
        
        Converting Numbers to Strings:
        num = 456
        str_num = str(num)  # Converts integer to string
        print(str_num, type(str_num))  # Output: "456" <class 'str'>

13. What is the difference between for and while loops ?
  - In Python, both for and while loops are used to repeatedly execute a block of code. However, they are designed for different purposes and operate in distinct ways.

    Feature	for Loop	while Loop

Usage:

for: Used when the number of iterations is known or when iterating over a sequence (e.g., list, string, range).
while: Used when the number of iterations is unknown and depends on a condition being met.

Condition Type
for: Iterates over a predefined sequence or range.
while: Executes as long as a specified condition is True.

Structure
for: Automatically handles iteration using an iterable.
while: Requires manual management of the loop variable and condition.

Termination
for: Stops when the sequence is exhausted.
while: Stops when the condition becomes False.

Typical Use Cases
for: Iterating through elements of a collection (e.g., list, dictionary) or running a loop a fixed number of times.
while: Repeating tasks until a specific condition is met (e.g., user input or reaching a threshold).

Infinite Loop Risk
for: Less prone to infinite loops as the iteration is predefined.
while: Higher risk of infinite loops if the condition is not updated properly.

```python
#6 Write a function to demonstrate the behavior of mutable and immutable arguments
def modify_arguments(mutable_arg, immutable_arg):
    print("Before modification:")
    print("Mutable argument:", mutable_arg)  # Will show the original mutable object
    print("Immutable argument:", immutable_arg)  # Will show the original immutable object

    # Modifying the mutable argument (list)
    mutable_arg.append(100)  # This will modify the original list

    # Modifying the immutable argument (integer)
    immutable_arg += 10  # This creates a new integer, does not modify the original one

    print("\nAfter modification:")
    print("Mutable argument:", mutable_arg)  # The list will be modified
    print("Immutable argument:", immutable_arg)  # The integer remains unchanged outside the function

# Creating mutable and immutable objects
my_list = [1, 2, 3]
my_integer = 50

# Calling the function
modify_arguments(my_list, my_integer)

# Checking the values after function call
print("\nOutside the function:")
print("Original list:", my_list)  # The list will have been modified
print("Original integer:", my_integer)  # The integer remains the same
```

```python
#8 Write a program to demonstrate the use of logical operators.
# Define some example variables
x = 10
y = 20
z = 30

# Using `and` operator (Both conditions must be True)
print("Using 'and' operator:")
if x < y and y < z:
    print("Both conditions are True: x < y and y < z")
else:
    print("One or both conditions are False.")

# Using `or` operator (At least one condition must be True)
print("\nUsing 'or' operator:")
if x > y or y < z:
    print("At least one condition is True: x > y or y < z")
else:
    print("Both conditions are False.")

# Using `not` operator (Negates the condition)
print("\nUsing 'not' operator:")
if not(x > y):
    print("The condition x > y is False, so 'not' makes it True.")
else:
    print("The condition x > y is True, so 'not' makes it False.")

# Combining all logical operators
print("\nCombining all logical operators:")
if (x < y and y < z) or not(x > y):
    print("Combined logical condition is True!")
else:
    print("Combined logical condition is False.")
```

```python
#10 Write code to demonstrate type casting with list elements.

# Original list with elements of different types (strings and floats)
mixed_list = ["10", "20.5", "30", "40.7", "50"]

print("Original mixed list:")
print(mixed_list)

# Type casting elements in the list to integers
int_list = [int(x) if '.' not in x else int(float(x)) for x in mixed_list]

# Type casting elements in the list to floats
float_list = [float(x) if '.' in x else float(x) for x in mixed_list]

# Type casting elements in the list to strings
string_list = [str(x) for x in mixed_list]

# Printing the results after type casting
print("\nList with elements cast to integers:")
print(int_list)

print("\nList with elements cast to floats:")
print(float_list)

print("\nList with elements cast to strings:")
print(string_list)

```
